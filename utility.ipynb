```python
#Checking for the fuel_type_codes
df_fuel_type_code = list(data['fuel_type_code_pudl'])
fuel_type_code_list =[]
for fuel_type_code in df_fuel_type_code:
    if df_fuel_type_code.count(fuel_type_code) > 1:
        if fuel_type_code not in fuel_type_code_list:
            fuel_type_code_list.append(fuel_type_code)
print(fuel_type_code_list)
```
